# Styles, and Layouts

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
```

```python
plt.style.use('ggplot')
data = np.random.randn(10)
# data
plt.plot(data)
plt.show()
```

```python
# plt.style.use('classic')
plt.style.use('default')
data = np.random.randn(10)
# data
plt.plot(data)
plt.show()
```

```python
plt.plot(data*-0.2)
plt.show()
```

```python
n =3
data = np.linspace(0,2*n*np.pi, 300)
def sinusoidal(sty):
    plt.style.use(sty)
    fig, ax = plt.subplots()
    ax.plot(data,np.sin(data), label= 'sine')
    ax.plot(data,np.cos(data), label= 'cos')
    ax.legend()
sinusoidal('default')
plt.show()
 
    
```

```python
sinusoidal('ggplot')
plt.show()
```

```python
sinusoidal('seaborn-v0_8-dark')
plt.show()
```

```python
with plt.style.context("Solarize_Light2"):
    data = np.linspace(0,6*np.pi)
    plt.plot(np.sin(data), 'g--')
    plt.show()
```

## Layouts

```python
plt.style.use('default')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, constrained_layout= True)
plt.show()
```

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure(constrained_layout = True)
specs = gridspec.GridSpec(nrows=2, ncols=2, figure=fig)
ax1 = fig.add_subplot(specs[0,0])
ax2 = fig.add_subplot(specs[0,1])
ax3 = fig.add_subplot(specs[1,0])
ax4 = fig.add_subplot(specs[1,1])
plt.show()
```
