## lower traingular matrix
- **where the diagonal and all the elements below the diagonal are 1 and the rest of the elements are 0**

```python
lower_triangle = np.tri(3,3, dtype=np.int8)
lower_triangle
```

### using tril method to create the triangluar matrix

```python
create_one = np.ones((3,3), dtype=np.int8)
alter_lower_tri = np.tril(create_one, k=-1)
alter_lower_tri
```

### upper triangular matrix

```python
upper_tri = np.triu(create_one)
upper_tri
```

### plotting ndarray routines

```python
x = np.arange(0,6,2, dtype=np.int8)
x
```

```python
y = x+1
y
```

```python
plt.plot(x,y, 'o--')
plt.show()
```

### Drawing multiple graphs

```python
plt.plot(x,y, 'o--')
plt.plot(x,-y, 'o--')
plt.show()
```

### adding the title

```python
plt.plot(x,y, 'o--')
plt.plot(x,-y, 'o--')
plt.title('y=x and y=-x')
plt.show()
```

### linspace

```python
N = 16
x = np.linspace(0,15,N, dtype=np.int8)
x
```

```python
y=x
```

```python
plt.plot(x,y,'o--')
plt.axis('off')
plt.show()
```

```python
y = np.logspace(0.1, 2,N)
y
```

```python
plt.plot(x,y,'o--')
plt.show()
```

### plotting in geometric progression

```python
y = np.geomspace(0.1, 2000, N)
y
```

```python
plt.plot(x,y, 'o--')
plt.show()
```
