### add title and label for the axis

```python
x = np.arange(3)
plt.plot(x, -x**2, x,-x**3, x, -2*x, x, -2**x)
plt.grid(True)
plt.xlabel(" x= np.arange(3)")
plt.xlim([0,2])
plt.ylabel(" y = f(x)")
plt.ylim([-8,0])
plt.title("Sample demo plot")
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, label = '-x**2')
plt.plot(x,-x**3, label = '-x**3')
plt.plot( x, -2*x, label = '-2*x')
plt.plot( x, -2**x, label = '-2**x')
plt.legend()
plt.grid(True)
plt.xlabel(" x= np.arange(3)")
plt.xlim([0,2])
plt.ylabel(" y = f(x)")
plt.ylim([-8,0])
plt.title("Sample demo plot")
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x,-x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2', '-x**3', '-2*x', '-2**x'])
plt.grid(True)
plt.xlabel(" x= np.arange(3)")
plt.xlim([0,2])
plt.ylabel(" y = f(x)")
plt.ylim([-8,0])
plt.title("Sample demo plot")
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x,-x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2', '-x**3', '-2*x', '-2**x'], loc ='center left')
plt.grid(True)
plt.xlabel(" x= np.arange(3)")
plt.xlim([0,2])
plt.ylabel(" y = f(x)")
plt.ylim([-8,0])
plt.title("Sample demo plot")
plt.show()
```

```python
x = np.arange(3)
plt.plot(x, -x**2, x,-x**3, x, -2*x, x, -2**x)
plt.legend(['-x**2', '-x**3', '-2*x', '-2**x'], loc ='center left')
plt.grid(True)
plt.xlabel(" x= np.arange(3)")
plt.xlim([0,2])
plt.ylabel(" y = f(x)")
plt.ylim([-8,0])
plt.title("Sample demo plot")
plt.savefig('axis_label_title')
plt.show()
```

### Colors, Styles, and Markers

```python
x = np.arange(5)
x
```

```python
y = x
y
```

```python
plt.plot(x,y*2, 'g')
plt.plot(x,y/2, 'y')
plt.plot(x,y,'r')
plt.plot(x,y+0.4, 'c')
plt.plot(x,y-0.4,'k')
plt.plot(x,y-0.2,'m')
plt.plot(x,y-0.4,'k')
plt.plot(x,y-0.8,'w')
plt.plot(x,y-0.7,'b')
plt.show()
```

### let's write the logic as follows

```python
plt.plot(x,y*2, 'g',x,y/2, 'y', x,y,'r',x,y+0.4, 'c',x,y-0.4,'k', x,y-0.2,'m',x,y-0.4,'k', x,y-0.8,'w', x,y-0.7,'b')
plt.show()
```

```python
plt.plot(x,y*2, '--',x,y/2, '*', x,y,'+',x,y+0.4, '.',x,y-0.4,'1', x,y-0.2,'2',x,y-0.4,'-1', x,y-0.8,'3', x,y-0.7,'4')
plt.show()
```

### Object Oriented Plotting

```python
fig, ax = plt.subplots()
```

```python
fig
```

```python
ax.plot(x, -x*2, label = '-x*2')
ax.plot(x, -x*4, label = '-x*4')
ax.plot(x, -x**2, label = '-x**2')
ax.plot(x, -x**4, label = '-x**4')
ax.plot(x, -2*x, label = '-2*x')
ax.set_xlabel('x = np.arange(10)')
ax.set_ylabel('y=x')
ax.set_title('Sample demo plot')
ax.legend()
ax.grid(True)
plt.show()
```

```python
ax.plot(x, -x*2, label = '-x*2')
ax.plot(x, -x*4, label = '-x*4')
ax.plot(x, -x**2, label = '-x**2')
ax.plot(x, -x**4, label = '-x**4')
ax.plot(x, -2*x, label = '-2*x')
ax.set_xlabel('x = np.arange(10)')
ax.set_ylabel('y=x')
ax.set_title('Sample demo plot')
ax.legend()
ax.text(4, -53000, 'simple demo plot')
ax.grid(True)
plt.show()
```

### subplots

```python
x = np.arange(3)
x
```

```python
plt.subplots_adjust(wspace = 0.3, hspace = 0.4)
plt.subplot(3,3,1)
plt.plot(x, -x**2)
plt.subplot(3,3,2)
plt.plot(x, -x**3)
plt.subplot(3,3,3)
plt.plot(x, -2*x)
plt.subplot(3,3,4)
plt.plot(x, -2**x)
plt.savefig("sub_plots.png")
plt.show()
```
